# Matplotlib basics

#### Importing Matplotlib library

```python
import matplotlib.pyplot as plt
```

#### The plot() method of Matplotlib can take in paramteters ranging from pandas series to numpy arrays
#### If we only pass a single value, it will act as Y axis and the x axis will be automatically 0+1

```python
plt.plot([2,4,6,5,7])
```

##### Now we will plot a basic example

```python
salaries = [23000,25000,21000,20000,30000]
age = [20,24,40,60,65]

plt.plot(age,salaries)
```

#### There is another method plt.show() where we only see the plot directly

```python
plt.plot(age, salaries)
plt.show()
```

## numpy.arange() - numpy library method

```python
import numpy as np

nums = np.arange(5)
```

```python
plt.plot(nums, nums)
```

```python
plt.plot(nums, nums)
plt.plot(nums, nums * nums)
```

# matplotlib.pyplot.figure()
#### In Matplotlib, matplotlib.pyplot.figure() creates a new top-level container object that holds all plotting elements like subplots, titles, legends, and canvases. 
#### Creates a new figure or activate an existing figure
#### Everyline added after plt.figure() gets added to the plot

```python
plt.figure()
plt.plot(nums,nums)
plt.plot(nums, nums**3)
plt.plot(nums, nums*nums)
# /////////////////////////////
plt.figure()
plt.plot(nums, nums**3)
plt.plot(nums, nums*nums)
```
