```python
import pandas as pd

house = pd.read_csv('../data/kc_house_data.csv')
house.head()
```

```python
# To find the MIN we will use the MIN method
house.min()
```

```python
# To find the max we will use the MAX method
house.max()
```

```python
# There is a SUM function
house.sum()
```

```python
# Using NUMERIC_ONLY, we can actually specify that we want only the sum of numeric values
house.sum(numeric_only=True)
```

```python
# Mean gives us the average value

house.mean(numeric_only=True)
```

```python
# Median gives us the value in the middle
house.median(numeric_only=True)
```

```python
# Mode gives us the most recurring value
house.mode()
```

```python
# The DESCRIBE method gives us the descriptive statistics
house.describe()
```

##### We can see statistics for particular datatypes as well

```python
house.describe(include='object')
```

```python
house.describe(include=['object', 'int64'])
```
