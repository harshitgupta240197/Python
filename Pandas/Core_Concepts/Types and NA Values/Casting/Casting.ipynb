# Casting Types with astype()

```python
import pandas as pd

titanic = pd.read_csv('../../data/titanic.csv')
titanic.head()
```

```python
titanic.info()
```

## We will try to convert the 'age' column from Object to Float

```python
titanic[titanic['age'] == '?']
```

#### We will first replace the '?' in the age column with None

```python
titanic['age'].replace(['?'], [None], inplace=True)
```

#### Now first we will create a duplicate column of age but with float type

```python
titanic['age_converted'] = titanic['age'].astype(float)
```

```python
titanic.info()
```

#### Now we will convert the original age column to float

```python
titanic['age'] = titanic['age'].astype('float')
```

##### Now we can drop the 'age_converted' column since it is of no use

```python
titanic.drop('age_converted', axis = 1)
```
