# pandas.to_numeric() method
#### converts a Pandas Series, list, or 1-D array containing non-numeric data types (like strings or objects) into a proper integer or floating-point format. It is the preferred method for data cleaning and validation over .astype() because it provides robust handling for dirty or missing data.

```python
import pandas as pd

titanic = pd.read_csv('../../data/titanic.csv')
titanic.head()
```

```python
titanic.info()
```

```python
titanic['age'] = pd.to_numeric(titanic['age'], errors='coerce')
```
