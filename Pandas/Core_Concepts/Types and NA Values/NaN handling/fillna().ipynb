# fillna()
#### Helps to replace NaN values with some other values

```python
import pandas as pd

games = pd.read_csv('../../../data/game_stats.csv')
```

```python
games
```

#### We will change all the NaN values with 0

```python
games.fillna(0)
```

#### We will target a specific column to replace NaN values

```python
games['league'].fillna('Batman')
```

#### We can target multiple columns now to replace NaN values using a dictionary

```python
games.fillna({'league' : 'Superman', 'points' : 0, 'assists' : 5})
```

### We can use fillna() to fill NaN values with values from another column of our choice

```python
import pandas as pd

sales = pd.read_csv('../../../data/sales.csv')
sales
```

```python
sales['shipping_zip'].fillna(sales['billing_zip'], inplace=True)
```
