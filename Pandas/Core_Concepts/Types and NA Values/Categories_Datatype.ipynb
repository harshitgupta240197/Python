# Categories Datatype
#### The category data type in pandas is a special hybrid data type designed to handle variables that take on a fixed and limited number of possible values (e.g., gender, blood types, or survey ratings).
#### Also helps tp reduce the memory usage if we shift an object to category

```python
import pandas as pd

titanic = pd.read_csv('../../data/titanic.csv')
titanic.head()
```

### We will convert the 'sex' column to Category type

```python
titanic['sex'] = titanic['sex'].astype('category')
```

```python
titanic.info()
```

```python
titanic['embarked'].value_counts()
```

```python
titanic['embarked'] = titanic['embarked'].astype('category')
```

#### We can also see significant memory usage reduction

```python
titanic.info()
```
