# .plot() from Series

```python
import pandas as pd

housing = pd.read_csv('../data/kc_house_data.csv')
housing.head()
```

```python
housing['bedrooms'].value_counts()
```

```python
housing['bedrooms'].value_counts().plot()
```

## Use the KIND= parameter to change the type of the plot

```python
housing['bedrooms'].value_counts().plot(kind='bar')
```

```python
import pandas as pd

titanic = pd.read_csv('../data/titanic.csv')
titanic.head()
```

# .plot() from Dataframes

```python
housing[['bedrooms', 'price']].plot(kind='scatter', x='bedrooms', y='price')
```
