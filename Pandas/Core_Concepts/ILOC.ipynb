# iloc[ ]
#### It is a purely integer-location based indexing for selection by position
## Basically iloc is position based and loc is label based

```python
import pandas as pd 

countries = pd.read_csv('../data/world-happiness-report-2021.csv', index_col=0)
countries.head()
```

```python
countries.iloc[[10,20,30]]
```

```python
countries.iloc[10:15]
```

### We can modify the dataframe columns as well

```python
countries.iloc[20:25, [0,1,2,3]]
```
