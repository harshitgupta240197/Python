# Modifying Dataframes

## 1. Dropping Columns

```python
import pandas as pd

btc = pd.read_csv('../data/coin_Bitcoin.csv')
btc.head()
```

##### Since the columns Symbol and Name are not of any help so we will drop them
##### The axis = 1 is for specifying column. axis = 0 is for rows
##### We can also specify by axis = 'columns'
##### This will not change the original dataset. For modifying the original dataset we need (inplace = True)

```python
btc_modified = btc.drop(labels='Name', axis=1)
```

```python
btc_modified.head()
```

##### Now we will drop the Symbol column and we will make changes to the original source this time using (inplace = True)

```python
btc.drop(labels='Symbol', axis=1, inplace=True)
```

```python
btc.head()
```

#### Multi-Column Drop

```python
btc.drop(labels=['SNo', 'Name', 'Volume'], axis=1).head()
```

##### Another way of Multi-column delete is by using (columns = [ ])

```python
btc.drop(columns=['SNo', 'Name', 'Volume']).head()
```

#### Another method involves no column deletion instead do selective selection of columns

```python
btc[['Date', 'High', 'Low']].head()
```

## 2. Dropping Rows

```python
import pandas as pd

countries = pd.read_csv('../data/world-happiness-report-2021.csv')
countries.head()
```

##### We will make a few modifications to the dataframe

```python
countries.set_index('Country name', inplace=True)
```

```python
countries.head()
```

##### Now we will remove the column containing the Country name 'Denmark'

```python
countries.drop(labels='Denmark', axis=0)
```

##### Another way of removing rows: 
##### We do not need to specify the axis = 0 since the axis is by default set to 0.

```python
countries.drop(['Finland', 'Iceland', 'Norway'])
```

#### Deleting rows by Index position

```python
btc.head(8)
```

```python
btc.drop(btc.index[2:5])
```
