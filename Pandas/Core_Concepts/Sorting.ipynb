```python
import pandas as pd

countries = pd.read_csv('../data/world-happiness-report-2021.csv')
countries.head()
```

# Sort

#### sort_values()
##### Whichever column we want to use to sort the entire dataframe sort_values(desired_column_to_sort)
##### The sorting occurrs in ascending order by default

```python
countries.sort_values('Healthy life expectancy', ascending=False).head(5)
```

#### inplace = True - Setting this parameter as true makes modification to the original object directly instead of returning a new row

```python
countries.sort_values("Healthy life expectancy", ascending=False, inplace=True)
```

## Sorting using Multiple Columns

```python
# Import required datasets first

import pandas as pd

houses = pd.read_csv('../data/kc_house_data.csv')
titanic = pd.read_csv('../data/titanic.csv')
```

```python
houses.sort_values(['price', 'bedrooms'], ascending=False)
```

```python
houses.sort_values(['bedrooms', 'bathrooms'], ascending=False)
```

## String level Sorting
#### Normal string sorting occurrs in this order ABCD....Z ->-> abcd...z

```python
titanic.sort_values('name', ascending=True)
```

### Lambda

#### To avoid the normal string sort ABCD....Z ->-> abcd...z we use the LAMBDA FUNCTION

```python
titanic.sort_values('name', key=lambda col: col.str.lower())
```
