# Bitwise Negation (~)
##### (Boolean Negation)
##### Basically it is like a flip switch
##### What used to be True will be False and what used to be False will be True

```python
titanic.head()
```

```python
women_on_titanic = titanic['sex'] == 'female'
titanic[women_on_titanic]
```

#### Once we use ~ then the switch will shift to Male

```python
titanic[~women_on_titanic]
```

##### Find homes that were built after 2010 or were renovated after 2010

```python
recent_houses = (houses['yr_built'] >= 2010) | (houses['yr_renovated'] >= 2010)
houses[recent_houses]
```

##### Now we will use ~ to flip and see the opposite of recent houses which is the old houses

```python
old_houses = houses[~recent_houses]
old_houses
```
