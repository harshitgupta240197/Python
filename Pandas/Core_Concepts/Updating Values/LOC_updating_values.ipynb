# Updating values using loc[ ]

```python
import pandas as pd

countries = pd.read_csv('../../data/world-happiness-report-2021.csv', index_col='Country name')
countries.head()
```

#### We can select what all columns that we want

```python
countries.loc['Denmark', ['Ladder score', 'Healthy life expectancy']]
```

```python
countries.loc[['Denmark', 'Switzerland', 'Iceland'], ['Ladder score', 'Healthy life expectancy']]
```

## We will now update values using loc[ ]

```python
countries.loc[['Denmark', 'Switzerland', 'Iceland'], ['Regional indicator']] = 'Scandinavia'
```

```python
countries.head(8)
```

## dataframe_name.loc[[Index labels], [column names]] = [Values to replace with in that row for the particular columns]

```python
countries.loc[['Finland', 'Denmark', 'Norway'], ['Regional indicator', 'Ladder score', 'upperwhisker']] = ['New Nation', 202, 102]
```

```python
countries.head(8)
```

# Boolean Array loc[ ] = Boolean Mask
#### Access a group of rows and columns by a boolean array.
#### .loc[ ] is primarily label based but may also be used with boolean array
## So basically .loc[ ] can help us update values in a dataframe using a condition

```python
import pandas as pd

houses = pd.read_csv('../../data/kc_house_data.csv')
houses.head()
```

#### First we will create a condition and store it inside a variable

```python
many_bedrooms = houses['bedrooms'] >= 10
```

#### Now we will access this boolean condition using .loc[ ]

```python
houses.loc[many_bedrooms]
```

#### Now we will perform selective column selection

```python
houses.loc[many_bedrooms, ['id', 'bedrooms', 'bathrooms', 'floors']]
```

#### Now to these selected columns with the applied condition we will perform updating of values

```python
houses.loc[many_bedrooms, ['bedrooms', 'bathrooms']] = ['too many', 10]
```

```python
houses.loc[many_bedrooms]
```

## Design a column for luxury and add it to the dataframe

```python
houses['Luxury Level'] = 'Normal'
```

```python
houses
```

#### We will create two conditions for luxury

```python
is_waterfront = houses['waterfront'] == 1
high_quality = houses['grade'] > 5
```

```python
houses.loc[is_waterfront & high_quality, ['Luxury Level']] = 'Luxury'
```

```python
houses.loc[is_waterfront & high_quality]
```
