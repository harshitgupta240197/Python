# Replacing Values

```python
import pandas as pd 

titanic = pd.read_csv('../../data/titanic.csv')
titanic.head()
```

##### Replace 'female' with 'f' in the 'sex' columns

```python
titanic['sex'].replace('female', 'F')
```

##### Replace both 'Male' and 'Female' with their first letters in the sex column

```python
titanic['sex'].replace(['female', 'male'], ['F', 'M'], inplace=True)
```

```python
titanic.head()
```

## If you want to repace something with NONE then you have to wrap NONE In square brackets like a list = [None]

#### Replace the '?' in the age column with None

```python
titanic['age'].tail()
```

```python
titanic['age'].replace(['?'], [None])
```

```python
titanic['age'].replace(['?'], [None], inplace=True)
```

```python
titanic['age'].value_counts(dropna=False)
```
