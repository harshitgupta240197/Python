# LOC -  .loc[ ]
#### loc helps us to access data by rows
#### .loc[ ] is primarily label based, but may also be used with a boolean array
#### Access a group of rows and columns by label(s) or a boolean array

```python
import pandas as pd 

countries = pd.read_csv('../data/world-happiness-report-2021.csv', index_col=0)
countries.head()
```

##### This will give a series

```python
countries.loc['Denmark']
```

##### This will give a dataframe

```python
countries.loc[['Denmark']]
```

```python
countries.loc[['Canada']]
```

#### Multiple rows from a dataframe

```python
countries.loc[['Canada', 'Yemen', 'Iceland']]
```

## .loc[ ] using Index numbers

```python
import pandas as pd

titanic = pd.read_csv('../data/titanic.csv')
titanic.head()
```

```python
titanic.loc[[6,7,8,12]]
```

#### We can achieve the same thing with a range as well

```python
titanic.loc[3:11]
```

#### We can also add a STEP to the range

```python
titanic.loc[15:20:2]
```

### We can edit the columns of the dataframe as well

```python
titanic.loc[20:25, ['name', 'age', 'survived']]
```
