# DateTime Components (dt)

```python
import pandas as pd

ufo = pd.read_csv('../../data/nuforc_reports.csv 19-47-12-819.csv')
```

```python
ufo.info()
```

#### First we will convert the date_time column to datetime

```python
ufo['date_time'] = pd.to_datetime(ufo['date_time'])
```

```python
ufo.info()
```

### Now we will use the YEAR component of DT

```python
ufo['date_time'].dt.year
```

```python
ufo['date_time'].dt.year.value_counts().head(10).plot(kind='bar')
```

### Now we will use the MONTH component of DT

```python
ufo['date_time'].dt.month
```

```python
ufo['date_time'].dt.month.value_counts().head(12).plot(kind='bar')
```

```python
ufo['date_time'].dt.day.value_counts().head(30).plot(kind='bar')
```

```python
ufo['date_time'].dt.hour.value_counts().head(24).plot(kind='bar')
```

### Day of the week - (Monday = 0 and Sunday = 6)

```python
ufo['date_time'].dt.day_of_week.value_counts().head(7).plot(kind='bar')
```
