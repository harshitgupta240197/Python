# Comparing Dates

```python
import pandas as pd

ufo = pd.read_csv('../../data/nuforc_reports.csv 19-47-12-819.csv')
```

```python
ufo.head()
```

```python
ufo['date_time'] = pd.to_datetime(ufo['date_time'])
ufo.info()
```

```python
ufo['date_time'].dt.year
```

### We can use comparison operators to create conditions

```python
ufo['date_time'] > '1980'
```

```python
ufo[ufo['date_time'] > '1980'].head()
```

```python
ufo[ufo['date_time'].dt.hour == 2.0].head()
```

```python
ufo[ufo['date_time'].between('2000', '2001')].sort_values('date_time').head()
```
