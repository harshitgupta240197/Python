# pd.to_datetime()
### pandas.to_datetime() is the primary function in the pandas library used to convert strings, lists, arrays, or data columns into official pandas datetime objects. 

```python
import pandas as pd

ufo = pd.read_csv('../../data/nuforc_reports.csv')
ufo.head()
```

```python
pd.to_datetime(ufo['date_time'])
```

### We will now use the to_datetime() method

```python
ufo['date_time'] = pd.to_datetime(ufo['date_time'])
```

```python
ufo.info()
```

## We can use the parse_dates = [ ] function to convert columns to date while importing the dataset

```python
pd.read_csv('../../data/nuforc_reports.csv').info()
```

```python
pd.read_csv('../../data/nuforc_reports.csv', parse_dates=['date_time']).info()
```
