# Date Math - Time Delta

```python
import pandas as pd

ufo = pd.read_csv('../../data/nuforc_reports.csv 19-47-12-819.csv')
```

```python
ufo.head()
```

### we will convert the 'posted' and 'date_time' column to date_time format

```python
ufo['posted'] = pd.to_datetime(ufo['posted'])
```

```python
ufo['date_time'] = pd.to_datetime(ufo['date_time'])
```

```python
ufo.info()
```

## When we perform this calculation (posted - date_time) the result that we get is called TIMEDELTA format

```python
ufo['time_taken_to_report'] = ufo['posted'] - ufo['date_time']
```
