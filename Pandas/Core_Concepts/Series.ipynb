# Series

## DROPNA parameter
##### By default in a nunique() count the NAN are left out but using (dropna = False) we can make NAN values count as well

```python
import pandas as pd

netflix = pd.read_csv('../data/netflix_titles.csv', sep='|')
netflix.head()
```

```python
netflix.info()
```

```python
# NUNIQUE without NAN values
netflix['director'].nunique()
```

```python
netflix['director'].nunique(dropna=False)
```

## nlargest() & nsamllest() methods of Series
##### It gives the n (n by default is 5) number of largest or smallest values in a series

```python
import pandas as pd

housing = pd.read_csv('../data/kc_house_data.csv')
housing.head()
```

```python
# In case we want to know the top 5 house prices
housing['price'].nlargest()
```

```python
# In case we want to know the top 10 house prices
housing['price'].nlargest(10)
```

```python
# In case we want to know the 10 lowest house prices in the dataset
housing['price'].nsmallest(10)
```

##### keep = 'all' mmethod helps to deal with duplicate values when seeing nlargest or nsmallest

```python
titanic['pclass'].nlargest(5)
```

```python
titanic['pclass'].nlargest(5, keep='all')
```

## nlargest() & nsmallest() in a dataframe

```python
# We are filtering the dataframe to show us the TOP 10 houses by PRICES
housing.nlargest(10, ['price'])
```

```python
#  Top 10 cheapest houses in the housing dataframe
housing.nsmallest(10, ['price'])
```

```python
# Find which homes have the largest number of bedrooms
housing.nlargest(10, ['bedrooms'])
```

##### Performing a Double Sort

```python
# In this case first we are sorting based on Bedrooms and then we are sorting on based on bathrooms
housing.nlargest(10, ['bedrooms', 'bathrooms'])
```
